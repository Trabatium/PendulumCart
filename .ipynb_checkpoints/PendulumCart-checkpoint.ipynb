# Pendulum on chart

## Includes

```python
%matplotlib inline
import numpy as np
import sympy as sp
from scipy.integrate import odeint

from matplotlib import pyplot as plt
from matplotlib import animation as ani
from IPython.display import HTML
```

## Init symbols and funtions

```python
t, m, M, l, g, Fb = sp.symbols('t m M l g Fb' )
x = sp.Function('x')(t)
theta = sp.Function('theta')(t)
F = sp.Function('F')(t)

# Generalized coordinates
z1 = sp.Function('z1')(t)
z2 = sp.Function('z2')(t)
z3 = sp.Function('z3')(t)
z4 = sp.Function('z4')(t)
states = [z1,z2,z3,z4]
u  = sp.Function('u')(t)
inputs = [u]
```

## Penulum on chart equations

Position of cart and mass

```python
xc = x
xm = x+l*sp.sin(theta)
ym = -l*sp.cos(theta)
```

Velocity of cart and mass

```python
dxc = xc.diff(t)
dxm = xm.diff(t)
dym = ym.diff(t)
```

Acceleration of cart and mass

```python
ddxc = dxc.diff(t)
ddxm = dxm.diff(t)
ddym = dym.diff(t)
```

Newtons law $\quad m \ddot{x}(t) = \sum_i F_i$

```python
eqCx = -m*ddxc + F + Fb*sp.sin(theta)
eqMx = -M*ddxm - Fb*sp.sin(theta)
eqMy = -M*ddym + Fb*sp.cos(theta) - M*g
```

Eliminate $F_b$ and obtain two equations of motion

```python
Fb_eqn = (eqMx/sp.sin(theta)+Fb).simplify()
EOM1 = eqCx.subs({Fb: Fb_eqn}).simplify()
EOM2 = (eqMx*sp.cos(theta)+eqMy*sp.sin(theta)).simplify()
```

## Generalized coordinates

```python
genCoords = {x:             z1, 
             x.diff(t):     z2, 
             theta:         z3, 
             theta.diff(t): z4, 
             F:              u}
sp.pprint(genCoords)
```

### Substitute into equations

```python
gEOM1 = EOM1.subs(genCoords)
gEOM1
```

```python
gEOM2 = EOM2.subs(genCoords)
gEOM2
```

## Obtain the form $\dot{z}(t) = f(z)$

Solve EOM2 for $\dot{z}_2$ and $\dot{z}_4$, 

```python
sol_dz4 = sp.solve(gEOM2,z4.diff(t))[0]
sol_dz4
```

```python
sol_dz2 = sp.solve(gEOM2,z2.diff(t))[0]
sol_dz2
```

Substitute $\dot{z}_2$ and $\dot{z}_4$ into EOM1

```python
EQ_dz2 = gEOM1.subs({z4.diff(t):sol_dz4}).simplify()
EQ_dz2 = sp.solve(EQ_dz2,z2.diff(t))[0]
EQ_dz2
```

```python
EQ_dz4 = gEOM1.subs({z2.diff(t):sol_dz2}).simplify()
EQ_dz4 = sp.solve(EQ_dz4,z4.diff(t))[0]
EQ_dz4
```

Get desired nonlinear vector valued function $f(z)$ such that $\dot{z} = f(z)$

```python
f = []
f.append(z2)
f.append(EQ_dz2)
f.append(z4)
f.append(EQ_dz4)
f
```

```python
class NLDiffEQN:
    # List of state variables
    states = None
    nStates = 0
    # List of input variables
    inputs = None
    nInputs = 0
    # Equilibrium point
    eqPoint = None
    
    # Dict of constants in equations
    constants = None
    
    # List of symbolic nonlinear equations
    equationsNL = None
    nEquations = 0
    # List of symbolic nonlinear equations (constants substituted)
    equationsNLc = None
    
    # A and B matrix of linear, time-varying, continuous system
    Acltv = None
    Bcltv = None
    Acltvc = None
    Bcltvc = None
    
    # Constructor
    def __init__(self,equations,states,inputs,eqPoint,constants = dict({})):
        self.equationsNL = equations
        self.equationsNLc = [0]*len(equations)
        self.nEquations = len(equations)
        if len(constants) != 0:
            for i, e in enumerate(equations):
                self.equationsNLc[i] = e.subs(constants)
        self.states = states
        self.nStates = len(states)
        self.inputs = inputs
        self.nInputs = len(inputs)
        self.eqPoint = eqPoint
        self.constants = constants
        
        # Compute continuous time, linear, time-varying system
        self.setLTI()
    
        
    def dfdtNL(self,zv,t,uv):
        res = [0]*len(zv)
        z = dict(zip(self.states,zv))
        u = dict(zip(self.inputs,uv))
        for i, e in enumerate(self.equationsNLc):
            res[i] = e.subs(z).subs(u).evalf()
        return res
    
    def setLTI(self):
        nz = self.nStates
        nu = self.nInputs
        # Init A and B matrix
        A = [[0 for x in range(nz)] for y in range(nz)]
        B = [[0 for x in range(nu)] for y in range(nz)]
        
    
        # Linearize system
        for i, eq in enumerate(self.equationsNL):
            # Populate A matrix
            for j, z in enumerate(self.states):
                A[i][j] = eq.diff(z).subs(self.constants)
            # Populate B matrix
            for j, u in enumerate(self.inputs):
                B[i][j] = eq.diff(u).subs(self.constants)
        
        # timeinvariant system
        self.Aclti = sp.Matrix(A).subs(self.eqPoint)
        self.Bclti = sp.Matrix(B).subs(self.eqPoint)
        
    
    def dfdtLTI(self,zv,t,uv):
        zd = dict(zip(self.states,zv))
        ud = dict(zip(self.inputs,uv))
        
        z = sp.Matrix(zv);
        u = sp.Matrix(uv);
        res = self.Aclti*z + self.Bclti*u
        return list(res.evalf())
```

```python
const = dict({M: 10,
              m: 1,
              g: 9.81,
              l: 1})
EQ_point = {z1: 0, z2: 0, z3: 0, z4: 0, u: 0}
z0 = [0,0,0.2,0]
u0 = [0]
Nt = 400
tmax = 5
t = np.linspace(0, tmax, Nt)
nldiff = NLDiffEQN(f,[z1, z2, z3, z4],[u],EQ_point,const)
```

```python
nldiff.dfdtLTI(z0, 0, u0)
```

```python
nldiff.dfdtNL(z0, 0, u0)
```

```python
resNL = odeint(nldiff.dfdtNL, z0, t, args=(u0,))
resLTI = odeint(nldiff.dfdtLTI, z0, t, args=(u0,))
res = [resNL, resLTI]
```

```python
def drawCartPendulum(i, data, col):
    # Chart position
    xc = data[i,0]
    yc = 0
    ax1.scatter(xc,yc,s=40,c=col)
    
    # Position of mass
    theta = data[i,2]
    xm = xc+np.sin(theta)
    ym = -np.cos(theta)
    ax1.scatter(xm,ym,s=150,c=col)
    
    # Draw line
    ax1.plot([xc,xm],[yc,ym],c=col)

def plotCartPosition(t, data, col, ax):
    xc = data[:,0]
    ax.plot(t, xc)

def plotBarAngle(t, data, col, ax):
    xc = data[:,2]
    ax.plot(t, xc)
    
def animate2(i,data):
    colors = ['k','b','r']
    ax1.clear()
    for k, d in enumerate(data):
        drawCartPendulum(i, d, colors[k])
    ax1.legend(("$Nonlinear$","Linear Time Invariant"))
    plt.xlim(-1, 1)
```

```python
fig2 = plt.figure(2)
for i, r in enumerate(res):
    ax2 = fig2.add_subplot(3, 1, 1)
    plotCartPosition(t,r,'k',ax2)
    ax2 = fig2.add_subplot(3, 1, 2)
    plotBarAngle(t,r,'k',ax2)
```
